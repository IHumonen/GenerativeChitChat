```python
import os

import torch

import youtokentome as yttm

from transformer import ChataboxModel
from utils import generate 
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')  
print(device)
```

```python
tokenizer_path = 'pretrained_bpe_lm.model'
tokenizer = yttm.BPE(model=tokenizer_path)
```

```python
chekpoint_timestamp = '2021-12-24-15-40-29'
epoch = 3
```

```python
folder = 'logs/checkpoint/'
```

```python
model_path = os.path.expanduser(f'{folder}{chekpoint_timestamp}_epoch_{str(epoch)}.pt')
```

```python
vocab_size = 30_000
emb_dim = 256
n_layers = 3
```

```python
max_len=25
```

```python
model = ChataboxModel(vocab_size, emb_dim, n_layers)
model.state_dict = torch.load(model_path)
```

```python
seed_texts = [
    'Спой мне песню',
    'Почему в продаже нет животного масла?',
    'Гнев, богиня, воспой Ахиллеса, Пелеева сына',
    'остекленение балкона 37 тыщ. р. (с установкой) это дорого?'
    'ребята, кстати к теме о любви! мне предложили руку и сердце! не знаю как быть?'
    'может ли сумбурное начало привести к логическому продолжению?)'
]

for text in seed_texts:
    print(f'Вопрос: {text}')
    print(f'Ответ: {generate(model, text, tokenizer, max_len=max_len, device=device)}')
```
